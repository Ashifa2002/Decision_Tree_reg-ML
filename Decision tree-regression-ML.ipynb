```python
import pandas as pd
```

```python
dataset=pd.read_csv("50_Startups.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']]
dependent=dataset[['Profit']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor(criterion='poisson')
regressor.fit(x_train,y_train)
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
y_pred=regressor.predict(x_test)
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
r_score
```

```python
import pickle  #saving the model using pickle libary
filename="Finalised_model_decisiontree.sav"
pickle.dump(regressor,open(filename,'wb'))
loaded_model=pickle.load(open("Finalised_model_decisiontree.sav",'rb')) #loading the model and predict using given inputs
result=loaded_model.predict([[1234,345,4565,1,0]])
result
```
